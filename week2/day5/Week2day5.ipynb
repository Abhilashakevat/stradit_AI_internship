```python
#t2
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print(df.columns)

# Fill missing Age values
df["Age"] = df["Age"].fillna(df["Age"].median())

# Create figure
plt.figure(figsize=(12, 8))

# Survival by class
plt.subplot(2, 2, 1)

class_survival = df.groupby("Pclass")["Survived"].mean()

plt.bar(class_survival.index.astype(str),class_survival.values)

plt.title("Survival by Class")
plt.xlabel("Pclass")
plt.ylabel("Survival Rate")

# Survival by sex
plt.subplot(2, 2, 2)

sex_survival = df.groupby("Sex")["Survived"].mean()

plt.bar(sex_survival.index,sex_survival.values)

plt.title("Survival by Sex")
plt.xlabel("Sex")
plt.ylabel("Survival Rate")


# Age histogram
plt.subplot(2, 2, 3)

plt.hist(df["Age"], bins=20, edgecolor="black")

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")


# Fare histogram
plt.subplot(2, 2, 4)

plt.hist(df["Fare"], bins=20, edgecolor="black")

plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")

# Adjust spacing
plt.tight_layout()

# Show dashboard
plt.show()
```

```python
#t3
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Fill missing Age values
df["Age"] = df["Age"].fillna(df["Age"].median())

# Separate survivor and non-survivor ages
survivors = df[df["Survived"] == 1]["Age"]

non_survivors = df[df["Survived"] == 0]["Age"]

# Histogram of survivors
plt.hist(survivors,bins=20,alpha=0.5,color="green",label="Survivors")

# Histogram of non-survivors
plt.hist(non_survivors,bins=20,alpha=0.5,color="red",label="Non-Survivors")

# Add title and labels
plt.title("Age Distribution: Survivors vs Non-Survivors")
plt.xlabel("Age")
plt.ylabel("Frequency")

# Show legend
plt.legend()

# Show graph
plt.show()
```

```python
#Block E
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)


# Fill missing Age values with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing Fare values with median
df["Fare"] = df["Fare"].fillna(df["Fare"].median())


# FamilySize = SibSp + Parch + 1 (including passenger)
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

# IsAlone = 1 if FamilySize == 1 else 0
df["IsAlone"] = np.where(df["FamilySize"] == 1, 1, 0)


plt.figure(figsize=(18, 5))



plt.subplot(1, 3, 1)

# Calculate survival rate by passenger class
survival_by_class = df.groupby("Pclass")["Survived"].mean()

# Create bar chart
plt.bar(survival_by_class.index.astype(str),survival_by_class.values,color=["gold", "silver", "saddlebrown"],edgecolor="black")

# Add title and labels
plt.title("Survival Rate by Pclass")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")


plt.subplot(1, 3, 2)

# Create histogram
plt.hist(df["Age"],bins=20,color="skyblue",edgecolor="black")

# Add title and labels
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")


plt.subplot(1, 3, 3)

# Create age bins (0-5, 5-10, ...)
bins = np.arange(0, 85, 5)

# Create age groups
df["AgeGroup"] = pd.cut(df["Age"], bins=bins)

# Calculate average fare by age group
fare_by_age = df.groupby("AgeGroup")["Fare"].mean()

# Plot line chart
plt.plot(fare_by_age.index.astype(str),fare_by_age.values,marker='o')

# Rotate labels for readability
plt.xticks(rotation=45)

# Add title and labels
plt.title("Average Fare by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Fare")


plt.tight_layout()


plt.savefig("week2_deliverable.png", dpi=150)


plt.show()

print("\nSUMMARY:\n")

print("1. First-class passengers survived more.")

print("2. Most passengers were young adults.")

print("3. Average fare changes with age groups.")

print("4. Missing values were cleaned.")

print("5. FamilySize and IsAlone columns were created.")
```
