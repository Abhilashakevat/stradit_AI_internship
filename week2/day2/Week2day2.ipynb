```python
#Week 2
#Day 2
#block A

def write_file(path,lines):
    with open(path,"w") as f:
        for line in lines:
            f.write(line+"\n")

lines=["apples","mangos","bananas"]
write_file("csv1.txt",lines)

#t2
def count_lines(path):
    try:
        cnt=0
        with open(path,"r") as file:
          for f in file:
               if f.strip()!= "":
                  cnt=cnt+1
          return cnt
    except FileNotFoundError:
          return 0

print("number of lines:",count_lines("csv1.txt"))

#t3
def count_words(path):
    try:
        cnt=0
        with open(path,"r") as file:
            for f in file:
                 words=f.split()
                 cnt=cnt+len(words)
            return cnt
    except FileNotFoundError:
        return 0
        
print("count of words:",count_words("csv2.txt"))    
#t4
def tail(path, n):
    try:
        with open(path, "r") as file:
            lines = file.readlines()
            return lines[-n:]
    except FileNotFoundError:
        return []


print(tail("csv1.txt", 2))

#Block B
#t1 doubt
with open("csv1.txt", "r") as file:
    lines = file.readlines()

with open("csv2.txt", "w") as f:
    for i, line in enumerate(lines, start=1):
        f.write(str(i) + ": " + line)

#t2
from datetime import datetime

def log(msg):
    with open("csv1.log", "a") as file:
        timestamp = datetime.now().isoformat()
        file.write(timestamp + " " + msg + "\n")
log("Program started")
log("User logged in")

#t3
def replace_in_file(path, old, new):
    with open(path, "r") as file:
        data = file.read()
    data = data.replace(old, new)
    with open(path, "w") as file:
        file.write(data)

replace_in_file("csv1.txt","mango","apple")




    
```

```python
#t2
import csv

with open("people.csv", "r") as file:

    reader = csv.DictReader(file)

    with open("adults.csv", "w", newline="") as new_file:

        writer = csv.DictWriter(new_file,fieldnames=["name", "age", "city"])
        writer.writeheader()
        for row in reader:
            if int(row["age"]) >= 30:
                writer.writerow(row)

print("adults.csv created")
```

```python
#t3
import csv

total = 0
count = 0

with open("students.csv", "r") as file:

    reader = csv.DictReader(file)

    for row in reader:

        try:
            total += int(row["score"])
            count += 1

        except ValueError:
            print("Invalid score:",row["score"])

if count > 0:
    average = total / count
    print("Average Score:",average)
else:
    print("No valid scores found")
```

```python
#t4
import csv

def csv_to_dict_list(path):

    data = []

    with open(path, "r") as file:

        reader = csv.DictReader(file)

        for row in reader:
            data.append(row)

    return data


result = csv_to_dict_list("people.csv")

print(result)
```

```python
#t5
import csv

def dict_list_to_csv(records, path):

    fields = records[0].keys()

    with open(path, "w", newline="") as file:

        writer = csv.DictWriter(file,fieldnames=fields)
        writer.writeheader()
        writer.writerows(records)

data = [
    {
        "name":"Abhi",
        "age":22,
        "city":"Mumbai"
    },

    {
        "name":"Rahul",
        "age":30,
        "city":"Pune"
    }
]

dict_list_to_csv(data, "output.csv")
```

```python
#Block 6
#t1
def safe_int(s, default=0):

    try:
        return int(s)

    except ValueError:
        return default


print(safe_int("50"))

print(safe_int("abc"))
```

```python
#t2

def open_file():

    while True:

        filename = input(
            "Enter file name: "
        )

        try:

            file = open(filename)

            print("Opened successfully")

            file.close()

            break

        except FileNotFoundError:

            print(
                "File not found, try again"
            )


open_file()
```

```python
#t3
def divide(a, b):

    try:

        return a / b

    except TypeError:

        print(
            "Enter numbers only"
        )

        return None

    except ZeroDivisionError:

        print(
            "Cannot divide by zero"
        )

        return None


print(divide(10,2))

print(divide(10,0))

print(divide(10,"a"))
```

```python
#t4

def read_numbers(path):

    skipped = 0

    with open(path, "r") as file:

        for line in file:

            try:

                number = int(line.strip())
                print(number)

            except ValueError:
                skipped += 1

    print(
        "Skipped:",
        skipped
    )


read_numbers(
"csv1.txt"
)
```

```python
#t5
class EmptyFileError(Exception):
    pass


def check_file(path):

    with open(path, "r") as file:

        content = file.read()
        if content == "":
            raise EmptyFileError("File is empty")
        print(content)
try:
    check_file(
        "csv3.txt"
    )

except EmptyFileError as e:

    print(e)
```

```python
#block E
import csv
try:

    with open("students1.csv","r") as file:
        reader = csv.DictReader(file)
        rows = []
        for index,row in enumerate(reader,start=1):
            total = 0
            subjects = ["math","science","english"]
            for subject in subjects:
                try:
                    value = int(row[subject])
                except ValueError:
                    print(f"Warning row {index}, column {subject}")
                    value = 0
                total += value
            average = total / 3
            row["total"] = total
            row["average"] = average
            rows.append(row)

    with open(
        "students_with_grades.csv",
        "w",
        newline=""
    ) as file:

        fields = ["name","math","science","english","total","average"]
        writer = csv.DictWriter(file,fieldnames=fields)
        writer.writeheader()
        writer.writerows(rows)
    print("File created successfully")

except FileNotFoundError:
 print("students1.csv not found")
```
