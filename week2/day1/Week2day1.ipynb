```python
#blok F

inventory = [
    {"sku": "A1", "name": "T-shirt", "price": 499, "tags": 
["clothing","cotton"], "stock": 12},
    {"sku": "B2", "name": "Mug",     "price": 199, "tags": 
["kitchen","ceramic"], "stock": 0},
    {"sku": "C3", "name": "Hoodie",  "price": 1299, "tags": 
["clothing","winter"], "stock": 4},
    {"sku": "D4", "name": "Notebook","price": 99,  "tags": ["stationery"],  
"stock": 50},
]

print(inventory)
print("only items with stock:")
for i in inventory:
    if i["stock"]>0:
        print(i)

print("most expensive item:")
max_price = max(item["price"] for item in inventory)
print(max_price)

avg_price = sum(item["price"]/len(item) for item in inventory)
print("average price:",avg_price)
unique_tags = []
#Print every unique tag used across all items.
for item in inventory:
    for tag in item["tags"]:
        if tag not in unique_tags:
            unique_tags.append(tag)

print(unique_tags)
# Build a dict mapping each tag to the list of items that have it.
tag_items = {}
for item in inventory:
    for tag in item["tags"]:
        if tag not in tag_items:
            tag_items[tag] = []
        tag_items[tag].append(item["name"])

print(tag_items)

# Add "on_sale": True to every item priced under 500.

for item in inventory:
    if item["price"] < 500:
        item["on_sale"] = True
print(inventory)
```
