```python
#Block C
import pandas as pd

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# 1. Show all female passengers in 1st class
print("Female Passengers in 1st Class:")
print(df[(df["Sex"] == "female") & (df["Pclass"] == 1)])


# 2. Show passengers who paid more than 100
print("\nPassengers with Fare > 100:")
print(df[df["Fare"] > 100])

# 3. Show passengers aged 20-30 who survived
print("\nPassengers aged 20-30 who survived:")
print(df[(df["Age"] >= 20) & (df["Age"] <= 30) & (df["Survived"] == 1)])

# 4. Show passengers whose name contains 'Master.'
print("\nPassengers whose name contains 'Master.':")
print(df[df["Name"].str.contains("Master.", regex=False)])

# 5. Show passengers travelling alone
print("\nPassengers travelling alone:")
print(df[(df["SibSp"] == 0) & (df["Parch"] == 0)])
```

```python
#Block D
import pandas as pd

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# 1. Average fare paid by survived vs not survived passengers
print("Average Fare by Survival Status:")
print(df.groupby("Survived")["Fare"].mean())


# 2. Average age per passenger class
print("\nAverage Age per Passenger Class:")
print(df.groupby("Pclass")["Age"].mean())


# 3. Median age per (Pclass, Sex) combination
print("\nMedian Age per (Pclass, Sex):")
print(df.groupby(["Pclass", "Sex"])["Age"].median())


# 4. Survival rate per embarkation port
print("\nSurvival Rate per Embarkation Port:")
print(df.groupby("Embarked")["Survived"].mean())


# 5. Maximum fare per class
print("\nMaximum Fare per Passenger Class:")
print(df.groupby("Pclass")["Fare"].max())
```

```python
#Block E
import pandas as pd

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Fill missing Age with median age
df["Age"].fillna(df["Age"].median(), inplace=True)

# Fill missing Embarked with most common value
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)

# Drop Cabin column
df.drop("Cabin", axis=1, inplace=True)

print("Missing values after cleaning:")
print(df.isnull().sum())

# 2. Create Title column from Name
df["Title"] = df["Name"].str.split(", ").str[1].str.split(".").str[0]
print("\nTitle Value Counts:")
print(df["Title"].value_counts())

# 3. Create FamilySize and IsAlone columns
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df["IsAlone"] = (df["FamilySize"] == 1).astype(int)
print("\nFamilySize and IsAlone Columns:")
print(df[["Name", "FamilySize", "IsAlone"]].head())

# 4. Create FareCategory using pd.cut
df["FareCategory"] = pd.cut(
    df["Fare"],
    bins=[0, 10, 50, 100, 600],
    labels=["low", "mid", "high", "premium"]
)
print("\nFare Category Counts:")
print(df["FareCategory"].value_counts())
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print("Original Shape:", df.shape)

# Fill missing Age with median
df["Age"].fillna(df["Age"].median(), inplace=True)

# Fill missing Embarked with most common port
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)

# Drop Cabin column
df.drop("Cabin", axis=1, inplace=True)

# Create Title column
df["Title"] = (df["Name"].str.split(", ").str[1].str.split(".").str[0])

# Create FamilySize
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

# Create IsAlone column
df["IsAlone"] = (df["FamilySize"] == 1).astype(int)

# Create FareCategory column
df["FareCategory"] = pd.cut(df["Fare"],bins=[0, 10, 50, 100, 600],labels=["low", "mid", "high", "premium"])

# Survival rate by Title
print("\nSurvival Rate by Title:")
print(df.groupby("Title")["Survived"].mean())

# Survival rate by FamilySize
print("\nSurvival Rate by Family Size:")
print(df.groupby("FamilySize")["Survived"].mean())

# Survival rate by FareCategory
print("\nSurvival Rate by Fare Category:")
print(df.groupby("FareCategory")["Survived"].mean())

df.to_csv("titanic_clean.csv", index=False)

print("\nCleaned dataset saved as titanic_clean.csv")


female_rate = df[df["Sex"] == "female"]["Survived"].mean()
male_rate = df[df["Sex"] == "male"]["Survived"].mean()


print(f"1. Females had a survival rate of {female_rate:.2f} while males had {male_rate:.2f}.")
print("2. Passengers paying higher fares generally had better survival chances.")
print("3. First-class passengers survived more often than lower-class passengers.")
print("4. Small families had better survival rates compared to very large families.")
print("5. Titles like 'Mrs' and 'Miss' showed higher survival rates than 'Mr'.")
```
