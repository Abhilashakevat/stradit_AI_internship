```python
#block F
#t1
import numpy as np

# Create fake image batch
imgs = np.random.randint( 0, 256, size=(8,32,32,3),dtype=np.uint8)

# 1. Shape and dtype
print("Shape:", imgs.shape)
print("Data type:", imgs.dtype)

# 2. Mean brightness of each image
brightness = imgs.mean(axis=(1,2,3))

print("Brightness of each image:")
print(brightness)

# 3. Mean of RGB channels
channel_mean = imgs.mean(axis=(0,1,2))

print("Channel mean:")
print(channel_mean)

# 4. Convert RGB to grayscale
gray = imgs.mean(axis=3)

print("Grayscale shape:")
print(gray.shape)

# 5. Standardisation
mean = imgs.mean(axis=(0,1,2))
std = imgs.std(axis=(0,1,2))
standard = (imgs - mean) / std

print("Mean after standardisation:")
print(standard.mean(axis=(0,1,2)))
print("Std after standardisation:")
print(standard.std(axis=(0,1,2)))

# 6. Find brightest image
brightest = brightness.argmax()
print("Brightest image index:")
print(brightest)
```
