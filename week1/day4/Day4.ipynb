```python
#for password single only for practice
from string import ascii_letters, digits, punctuation
import random

all_char=ascii_letters+digits+punctuation
p_len=int(input("Enter the length of password:\n"))

p_word=" "
for i in range(p_len):
    p_word=p_word+random.choice(all_char)
print("password is:",p_word)

if len(p_word) > p_len and p_word.upper() and p_word.lower():
    print("Strong Password")
else:
    print("Weak Password")
```

```python
#T3
from string import ascii_letters, digits, punctuation
import random

all_char = ascii_letters + digits + punctuation

def generate_password(length):
    password = ""
    for i in range(length):
        password += random.choice(all_char)
    return password


count = int(input("How many passwords you want:\n"))
length = int(input("Enter password length:\n"))

password_list = []

for i in range(count):
    password_list.append(generate_password(length))

print("Generated Passwords:")
for pwd in password_list:
    print(pwd)
```

```python
#T4
text = "Python is a powerful programming language"

vowel = ['a', 'e', 'i', 'o', 'u']

def text_analyse(text):
    cntv = 0
    cntc = 0
    cnts = 0
    cnt_ch = 0

    words = text.split()

    for ch in text:
        cnt_ch += 1

        if ch.lower() in vowel:
            cntv += 1

        elif ch.isalpha():
            cntc += 1

        elif ch == " ":
            cnts += 1

    cntw = len(words)

    # longest and shortest word
    longest_word = max(words, key=len)
    shortest_word = min(words, key=len)

    print("No of characters:", cnt_ch)
    print("No of vowels:", cntv)
    print("No of consonants:", cntc)
    print("No of spaces:", cnts)
    print("No of words:", cntw)
    print("Longest word:", longest_word, ":", len(longest_word))
    print("Shortest word:", shortest_word, ":", len(shortest_word))


text_analyse(text)
```

```python
def km_to_miles(km):
    return km * 0.621371

def kg_to_pounds(kg):
    return kg * 2.20462

def liters_to_gallons(l):
    return l * 0.264172


while True:
    print("\nChoices:")
    print("1. Km to Miles")
    print("2. Kg to Pounds")
    print("3. Liters to Gallons")
    print("4. Quit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        n = float(input("Enter km: "))
        result = km_to_miles(n)
        print(n, "km =", result, "miles")

    elif choice == 2:
        n = float(input("Enter kg: "))
        result = kg_to_pounds(n)
        print(n, "kg =", result, "pounds")

    elif choice == 3:
        n = float(input("Enter liters: "))
        result = liters_to_gallons(n)
        print(n, "liters =", result, "gallons")

    elif choice == 4:
        print("Exited")
        break

    else:
        print("Invalid choice")
```

```python
def get_student_info():
    name = input("Enter your name: ")
    marks = []

    for i in range(1, 7):
        m = int(input(f"Enter marks for subject {i}: "))
        marks.append(m)

    return name, marks


def calculate_state(marks):
    total = sum(marks)
    avg_marks = total // len(marks)
    high_marks = max(marks)
    min_marks = min(marks)

    return total, avg_marks, high_marks, min_marks


def determine_grade(avg_marks):
    if avg_marks >= 90:
        return "A+"
    elif avg_marks >= 80:
        return "A"
    elif avg_marks >= 70:
        return "B"
    elif avg_marks >= 60:
        return "C"
    else:
        return "F"


def get_report(name, marks, stats, grade):
    total, avg, high, low = stats

    print("\n===== STUDENT REPORT =====")
    print("Name:", name)

    print("\nSubject Marks:")
    for i in range(len(marks)):
        print(f"Subject {i+1}: {marks[i]}")

    print("\nTotal Marks:", total)
    print("Average Marks:", avg)
    print("Highest Marks:", high)
    print("Lowest Marks:", low)
    print("Grade:", grade)


# MAIN PROGRAM
name, marks = get_student_info()
stats = calculate_state(marks)
grade = determine_grade(stats[1])
get_report(name, marks, stats, grade)
```
