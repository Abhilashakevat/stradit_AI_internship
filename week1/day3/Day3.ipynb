```python
#Day3
#T1
age=int(input("Enter your age:\n"))
if age<=12 and age>0:
    print("category:child")
elif age>=13 and age<=17:
    print("category:Teenager")
elif age>=18 and age<=24:
    print("category:Young Adults")
elif age>=25 and age<=59:
    print("category:Adults")
elif age>=59 and age<=150:
    print("category:Senior")
else:
    print("invalid age")
```

```python
#T2
n1 = int(input("enter first no: "))
n2 = int(input("enter Second no: "))

while True:
    ope = input("Enter operation")

    if ope == "stop":
        print("Calculator stopped")
        break
    elif ope == "+":
        print("ADDITION IS:", n1 + n2)

    elif ope == "-":
        print("SUBSTRACTION IS:", n1 - n2)

    elif ope == "*":
        print("MULTIPLICATION IS:", n1 * n2)

    elif ope == "/":
        if n2 != 0:
            print("Division is:", n1 / n2)
        else:
            print("CANT DIVIDE BY ZERO")

    elif ope == "//":
        print("DIVISION IS:", n1 // n2)

    elif ope == "%":
        print("REMAINDER IS:", n1 % n2)

    elif ope == "**":
        print("EXPONENTIOATION IS:", n1 ** n2)

    else:
        print("Unknown operator")
        break
 
```

```python
#T6

p = input("Enter password: ")

digit = False
upper = False

for ch in p:
    if ch.isdigit():
        digit = True

    if ch.isupper():
        upper = True

if len(p) < 6:
    print("Weak password")

elif len(p) >= 6 and len(p) <= 9 and digit and upper:
    print("Medium password")

elif len(p) >= 10 and digit and upper:
    print("Strong password")

else:
    print("Password is not valid")
```
